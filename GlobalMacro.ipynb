La tasa de paro es un dato *retrasado* del ciclo económico. 
porque resulta que la propia bolsa es un indicador bastante adelantado del ciclo económico y que el empleo es un indicador retrasado.

Vemos que si la bolsa hace suelo (en octubre de 2002), la tasa de paro hará techo unos meses después ( junio de 2003), 9 meses después. Por lo tanto vemos que es un indicador retrasado.

Esto ocurre porque al comienzo de una crisis las empresas primero intentan no despedir a sus empleados, pero a medida que esta avanza, el despido se vuelve inevitable por el deterioro de las condiciones financieras de la empresa.

Lo que nos interesa es buscar indicadores lo más adelantados posibles, por lo que el dato de paro, al ser una información retrasada, no nos sirve.

Lo que nos interesa **como inversores** es observar los indicadores macroeconómicos **más adelantados** y que a la vez sean **sencillos** de interpretar.
