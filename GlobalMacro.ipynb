La tasa de paro es un dato *retrasado* del ciclo econmico. 
porque resulta que la propia bolsa es un indicador bastante adelantado del ciclo econmico y que el empleo es un indicador retrasado.

Vemos que si la bolsa hace suelo (en octubre de 2002), la tasa de paro har techo unos meses despus ( junio de 2003), 9 meses despus. Por lo tanto vemos que es un indicador retrasado.

Esto ocurre porque al comienzo de una crisis las empresas primero intentan no despedir a sus empleados, pero a medida que esta avanza, el despido se vuelve inevitable por el deterioro de las condiciones financieras de la empresa.

Lo que nos interesa es buscar indicadores lo ms adelantados posibles, por lo que el dato de paro, al ser una informacin retrasada, no nos sirve.

Lo que nos interesa **como inversores** es observar los indicadores macroeconmicos **ms adelantados** y que a la vez sean **sencillos** de interpretar.
